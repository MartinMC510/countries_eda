### Make the Cleaning Population Dataset Function

```python
def clean_pop(df):
    # assign passed in df to sample_df to handle the dynamic change
    sample_df = df
    # fill NaN values
    sample_df.fillna('fill',inplace=True)

    # add a respective year column where it is just the relevant year to the data frame

    sample_df['year'] =  [re.findall('\d{4}',str) for str in sample_df["Italy, regions, provinces  "] 
     # this above returned the year of the dataframe
     # with multiple empty list
     for year in re.findall('\d{4}',str) 
     # this second loop returned a one element list that element being of type list
        ][0][0] # extract the year
    
    # remove irrelvent values
    sample_df = sample_df.iloc[8:]
    
    # rename the columns
    sample_df.rename(columns={'Italy, regions, provinces  ':'region','Unnamed: 1':'male',
                          'Unnamed: 2':'female','Unnamed: 3':'region_pop'},inplace=True)
    # extract relavent columns
    sample_df = sample_df[['region','region_pop','year']]
    
    # remove white spaces from region values
    sample_df['region'] = [region.strip() for region in sample_df['region']]
    # cast total_pop as int
    sample_df.total_pop = sample_df.region_pop.astype(int)

    # cast year as int
    sample_df.year = sample_df.year.astype(int)
    
    # remove translated region names(i.e. a value that contains '/')
    sample_df['region'] = [re.split(' /',region)[0] for region in sample_df['region']]
    
    # remove duplicates that occured
    sample_df.drop_duplicates(inplace=True)
    
    # filter for the regions
    region_list = [sample_df.loc[sample_df.region == region] for region in regions_in_italy]
    # the expression above returned a list conating a the row of data for each region
    # this expression will have them all be on one data frame
    region_df = pd.concat(region_list).reset_index().drop('index',axis=1)
    
    
    # calculate the total population for of Italy
    total_pop_it = region_df.region_pop.sum()
    # add percent column
    region_df['percent_of_total_pop'] = [round(region_pop/total_pop_it*100,2) 
                                     for region_pop in region_df.region_pop]
    
    return region_df
    
```

```python
# pass all the datasets into the function
yearly_pop_region_ls = [clean_pop(df) for df in pops_by_region ]
# dataframe with total population counts for each in region in Italy through 2019-2024
yearly_pop_region = pd.concat(yearly_pop_region_ls).reset_index().drop('index',axis=1)
```

### The Relationship Between GDP and the Average Wage

##### Insight:
Following the points, there seems to be some correlation between the two variables. As the GDP went up, so did the average wage. Yet, in 2020, when the GDP dropped significantly, the average wage sustained its positive progress. The correlation coefficient suggests a moderate strength between these variables, meaning that as the GDP increases, the average wage increases as well. If the GDP hadn't tanked in 2020, would the correlation coefficient have been higher?

### Italy's Population Layers

```python
# facet by each year in the population data
fig,ax = plt.subplots(2,3,figsize=(10,6))

# store the time series for this data

pop_years = pd.unique(yearly_pop_region_geo.year)

# have the range of the years
range_pop_years = np.arange(len(pop_years))

# have a dataframe for each year
pop_by_year = [yearly_pop_region_geo.loc[yearly_pop_region_geo.year == year] for year in pop_years]

# plot the population density for each year
plots_pop = [pop_by_year[i].plot(column='percent_of_total_pop', cmap='copper',ax=ax[i//3][i%3]) 
 for i in range_pop_years]

# set the grid to be a shade of blue 
[plots_pop[i].set_facecolor('steelblue') for i in range_pop_years]

# add titleoblique
plt.suptitle('Population Density by Region for Italy',fontsize=16,fontstyle='oblique',fontweight='bold')

# add titles for each plot

[plots_pop[i].set_title(pop_years[i]) for i in range_pop_years ]

# color bar which accounts for all subplots

## create the range of the rates and normalize the data. using the min and max rates in the working dataframe
norm_pop = mcolors.Normalize(vmin=yearly_pop_region_geo.percent_of_total_pop.min(),
                         vmax=yearly_pop_region_geo.percent_of_total_pop.max())
## create a scalarmappable which will allow for the rates to be mapped to general color bar
mappable_pop = plt.cm.ScalarMappable(cmap='copper',norm=norm_pop)
## define the colorbar
cbar_pop = fig.colorbar(mappable_pop,ax=ax)
cbar_pop.set_label('Population (%)',fontsize=12)

 # removes the tick on both axes
plt.setp(ax, xticks=[], yticks=[])

plt.show()
```

##### Insight:
The Northern region has held about 16% of Italy's population in recent years. Also, the Northern region appears to have a higher population compared to certain regions in central and southern Italy. The population percentage has also remained consistent from 2019 to 2024.

### Italy's GDP Layers 

```python
# adding the percent at which a region contributes to the total GDP value for that year
 
## filter for each year and obtain the gdp value for each region that year 
gdp_for_year_by_region = [yearly_econ_region_geo.loc[yearly_econ_region_geo.year == year].GDP 
                          for year in total_gdp_by_year.year]

## divide the gdp value for each region over the total gdp value which respective to each year  
yearly_econ_region_geo['gdp_percent'] = pd.concat([round(gdp_for_year_by_region[i]/total_gdp_by_year.loc[i][1]*100,2)
         for i in np.arange(len(total_gdp_by_year.year))])

```

```python
# facet by each year in the econ data that have GDP values
fig,ax = plt.subplots(2,5,figsize=(12,6))

# store the time series for this data

econ_years = np.unique(yearly_econ_region.year)

# have the range of the years
range_econ_years = np.arange(len(econ_years))

# have a dataframe for each year
gdp_by_year = [yearly_econ_region_geo.loc[yearly_econ_region_geo.year == year] for year in econ_years]

# plot the population density for each year
## using a diffrent colormap here to represent GDP 
plots_gdp = [gdp_by_year[i].plot(column='gdp_percent', cmap='Greens',ax=ax[i//5][i%5]) 
 for i in range_econ_years]

# set the grid to be a shade of blue 
[plots_gdp[i].set_facecolor('steelblue') for i in range_econ_years]

# add title
plt.suptitle('Regional GDP Contribution for Italy',fontsize=18,fontstyle='oblique',fontweight='bold')

# add titles for each plot

[plots_gdp[i].set_title(econ_years[i]) for i in range_econ_years ]

# color bar which accounts for all subplots

## create the range of the rates and normalize the data. using the min and max rates in the working dataframe
norm_gdp = mcolors.Normalize(vmin=yearly_econ_region_geo.gdp_percent.min(),
                         vmax=yearly_econ_region_geo.gdp_percent.max())
## create a scalarmappable which will allow for the rates to be mapped to general color bar
mappable_gdp = plt.cm.ScalarMappable(cmap='Greens',norm=norm_gdp)
## define the colorbar
cbar_gdp = fig.colorbar(mappable_gdp,ax=ax)
cbar_gdp.set_label('GDP (%)',fontsize=12)

 # removes the tick on both axes
plt.setp(ax, xticks=[], yticks=[])

plt.show()
```

##### Insight:

The Northern region has consistently contributed more to Italy's GDP over the past decade, maintaining a share of around 5.5% or higher. In contrast, the Southern region has remained around 3-4%. One particular region in the South has shown an increase in its contribution, as seen in the 2022 projection.
